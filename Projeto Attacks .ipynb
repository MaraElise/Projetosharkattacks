```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
attacks =pd.read_csv('attacks.csv', sep=',', index_col = -1, encoding='latin-1')
```

```python
attacks
```

```python
attacks.columns
```

```python
attacks.rename(columns={'Fatal (Y/N)':'fatal'}, inplace=True)
```

```python
attacks.columns = [x.strip().lower().replace(' ','_') for x in attacks.columns]
attacks.columns
```

```python
attacks.dropna(how='all') 
```

```python
projshark = attacks[['date','year','country','location','activity','fatal','type','area']]

projshark.dropna(how='all') 
```

```python
projshark['country'].value_counts()
```

```python
projsharkattacks = projshark[(projshark["year"] >=2000)]

projsharkattacks.year.unique()
```

```python
proj_sharkattacks = projsharkattacks[(projsharkattacks["country"] =="USA") | (projsharkattacks["country"] =="AUSTRALIA")] 
proj_sharkattacks
```

```python
projsharkattacks['country'].value_counts()
```

```python
def tratameno_fatal(x):
    if x == 'Y':
        return 1
    elif x == 'N':
        return 0
    else:
        return 'Unknown'
```

```python
proj_sharkattacks = proj_sharkattacks.reset_index()
```

```python
proj_sharkattacks['fatal'].value_counts()
```

```python

proj_sharkattacks['occurrences'] = proj_sharkattacks['fatal'].map(tratameno_fatal)
proj_sharkattacks['qt']=1
proj_sharkattacks
```

```python
sharkattacksproj = proj_sharkattacks[proj_sharkattacks['occurrences'] != 'Unknown']
sharkattacksproj
```

```python
sharkattacksproj['area'].value_counts()
```

```python
sharkattacksproj['country'].value_counts()
```

```python
year_ = pd.qcut(sharkattacksproj.year, 3)

year_.unique()
```

```python
sharkattacksproj = sharkattacksproj.reset_index()
```

```python
names = ['2000 a 2007', '2008 a 2013', '2014 a 2018']
years = pd.qcut(sharkattacksproj.year, 3,labels=names) 
sharkattacksproj['years']=years
sharkattacksproj.groupby(['years','country',])[['fatal']].count()
```

```python
sharkattacks_perc = sharkattacksproj.groupby(['years','country'])[['occurrences']].count().sort_values(by=['years','occurrences'], ascending=False)
```

```python
shark_attacks_perc = round(sharkattacks_perc[['occurrences']]*100 / sharkattacks_perc.groupby('years')[['occurrences']].sum(),2).astype('str')+'%'
sharkattacks_perc['percente']=shark_attacks_perc['occurrences']
sharkattacks_perc
```

```python
sharkattacks_country = sharkattacksproj.groupby('country')[['occurrences']].count().sort_values(by=['occurrences'], ascending=False)
```

```python
sharkattacks_country
```

```python
sharkattacks_perc2 = sharkattacksproj.groupby(['years','fatal','country'])[['occurrences']].count().sort_values(by=['years','occurrences'], ascending=False)
```

```python
shark_attacks_perc2 = round(sharkattacks_perc2[['occurrences']]*100 / sharkattacks_perc2.groupby('years')[['occurrences']].sum(),2).astype('str')+'%'
sharkattacks_perc2['percente']=shark_attacks_perc2['occurrences']
sharkattacks_perc2
```
